2. Female to Male Ratio


MATCH(female:Person{gender:0}) WITH count(female) AS femaleCount  
MATCH(n) with count(n) as TotalCount, femaleCount  
RETURN (femaleCount*1.0f/TotalCount*1.0f) as FemaletoMaleRatio
